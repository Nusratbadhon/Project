```python
#1

import logging
import os
import csv

logging.basicConfig(filename= "word_count.log" ,level= logging.DEBUG ,format= '%(asctime)s %(levelname)s %(message)s')

class Text:
    def __init__(self,file):
        self.__file= file
        logging.info(f"Given file is: {self.__file}")
    def exception():
        if type(i)==int:
            raise TypeError("The type integer is not acceptable",x)
        return x
    
 
    def word(self):
        self.words=[]
        
        try:
            with open(self.__file,'r+',encoding='utf8') as f:
                file=csv.reader(f)
                for j in file:
                    if isinstance(j,int):
                         exception()
                    else:
                         self.words.append(j[0])
                            
            f.close()
                                                          
        except Exception as e:
            logging.exception("Handle this" + str(e)+"exception")
                
        logging.info(f"All the word list is : {self.words}")
        logging.info(f"{len(self.words)}")
                        
        return (self.words)
        
    def counts_words(self):
        try:
            counts = dict()
           
            for word in self.words:
                if word in counts:
                    counts[word] += 1
                else:
                    counts[word] = 1
            
            logging.info(f"{list(counts.items())}")

            return list(counts.items())
            
        except Exception as e:
            
            logging.error("Error occured!!!")
            logging.exception(e)
```

```python

import string

logging.basicConfig(filename= "alpha_count.log" ,level= logging.DEBUG ,format= '%(asctime)s %(levelname)s %(message)s')


class Alphabet(Text):

    logging.info(f"This is a child class of text class")
        
    def counts(self):
        
        amounts = dict.fromkeys(string.ascii_lowercase, 0)
        try:        
            
            for i in range(len(self.words)):
                    amounts[self.words[i][0].lower()] += 1 # Ignore case and increment letter counter
            return list(amounts.items())
        except Exception as e:
            logging.exception(f"This exception {e} has to be handled!!")
```

```python
import re
import logging
import os
import csv

class Strng(Text):

    
    def just_str(self): 
        new=[]
        for i in range(len(self.words)):  
            x= re.findall("[\d;@!#+=\?><"'%*''('')'"-.$%^&/_*]", self.words[i])
            s=''
            
            for j in self.words[i]:
                if j not in x:
                    s+=j
            if len(s)!=0:    
                new.append(s)
        return new
```
